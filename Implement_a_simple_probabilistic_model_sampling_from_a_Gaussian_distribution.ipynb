```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_sampling(mu=0, sigma=1, n_samples=1000, seed=None):
    if seed is not None:
        np.random.seed(seed)
    samples = np.random.normal(loc=mu, scale=sigma, size=n_samples)
    return samples

def demo():
    # Parameters
    mu = 5.0 # mean
    sigma = 2.0 # standard deviation
    n_samples = 5000 # number of points

    # Generate samples
    samples = gaussian_sampling(mu, sigma, n_samples, seed=42)

    # Print first few samples
    print("First 10 samples:", samples[:10])

    # Plot histogram
    plt.hist(samples, bins=50, density=True, alpha=0.6, color='blue', edgecolor='black')
    plt.title(f"Gaussian Distribution Sampling (μ={mu}, σ={sigma})")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.grid(True)
    plt.show()

demo()
```
